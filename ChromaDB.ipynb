```python
!pip install langchain-community chromadb sentence-transformers gradio

import os
from sentence_transformers import SentenceTransformer
from langchain_community.vectorstores import Chroma
from langchain.schema import Document  # Import Document class

# Modello di embedding
embedding_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

class LocalEmbeddingFunction:
    def embed_documents(self, texts):
        return embedding_model.encode(texts).tolist()

    def embed_query(self, query):
        # Return the inner list of floats instead of a list containing a list
        return embedding_model.encode([query]).tolist()[0]

embedding_function = LocalEmbeddingFunction()

# Inizializza ChromaDB
CHROMA_PATH = "/content/chroma_db"  # Cartella per la persistenza
db = Chroma(persist_directory=CHROMA_PATH, embedding_function=embedding_function)

# Funzione per caricare i file e popolare il database
def load_books_to_db(folder_path):
    if not os.path.exists(folder_path):
        print(f"Error: Directory '{folder_path}' not found.")
        return

    file_paths = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith(".txt")]
    if not file_paths:
        print(f"Error: No '.txt' files found in '{folder_path}'.")
        return

    documents = []
    for file_path in file_paths:
        try:
            with open(file_path, "r", encoding="utf-8") as file:
                content = file.read()
                # Create Document object with metadata
                documents.append(Document(page_content=content, metadata={"title": os.path.basename(file_path)}))
        except Exception as e:
            print(f"Error reading file '{file_path}': {e}")

    if documents:
        db.add_documents(documents)
        db.persist()
        print(f"Libri caricati e aggiunti al database! Totale: {len(documents)}")
    else:
        print("No documents were loaded to add to the database.")

# Funzione per eseguire una query
def query_database(query_text, top_k=3):
    try:
        results = db.similarity_search(query_text, k=top_k)
        if results:
            for idx, result in enumerate(results, 1):
                print(f"Risultato {idx}:")
                print(f"Contenuto: {result.page_content[:200]}...")  # Mostra i primi 200 caratteri
                print(f"Metadata: {result.metadata}")
                print("-" * 50)
        else:
            print("No results found for the query.")
    except Exception as e:
        print(f"Error during query execution: {e}")

# Percorso dei file
folder_path = "./libri"
load_books_to_db(folder_path)

# Esegui una query di esempio
query_text = "Chi è Beatrice nella Divina Commedia?"
query_database(query_text, top_k=3)
```

```python
# Installazione delle librerie (se necessario)
# pip install chromadb langchain sentence-transformers gradio

# Importa le librerie richieste
from sentence_transformers import SentenceTransformer
from langchain_chroma import Chroma
from langchain.prompts import ChatPromptTemplate
from langchain_huggingface import HuggingFaceEndpoint
from langchain.schema import AIMessage, HumanMessage
import gradio as gr

# TOKEN Hugging Face
HUGGING_FACE_TOKEN = "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"

# Configurazione Hugging Face API
repo_id = "Qwen/Qwen2.5-Coder-32B-Instruct"
PROMPT_TEMPLATE = """
Answer the question based on the context provided. If no relevant information is found, state so.

Context:
{context}

Question:
{question}

Answer:
"""

# Modello di embedding
embedding_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

class LocalEmbeddingFunction:
    def embed_documents(self, texts):
        embeddings = embedding_model.encode(texts)
        return embeddings.tolist()

    def embed_query(self, query):
        query_embedding = embedding_model.encode(query)
        return query_embedding.tolist()

embedding_function = LocalEmbeddingFunction()

# Inizializza Chroma DB
CHROMA_PATH = "./chroma_db"  # Directory per salvare il database
db = Chroma(persist_directory=CHROMA_PATH, embedding_function=embedding_function)

# Classe per Hugging Face API
class LLM:
    def __init__(self):
        self.llm = HuggingFaceEndpoint(
            repo_id=repo_id,
            huggingface_token=HUGGING_FACE_TOKEN,
            temperature=0.2,
        )

    def generate_response(self, prompt):
        return self.llm.invoke(prompt)

# Funzione per evitare duplicati
def add_document_to_chroma(content, metadata, db):
    """
    Aggiunge un documento al database Chroma se non è già presente.
    """
    # Cerca documenti simili nel database
    existing_docs = db.similarity_search(content, k=1)
    if existing_docs and existing_docs[0].page_content.strip() == content.strip():
        print(f"Documento duplicato trovato: {metadata['title']}")
        return  # Non aggiungere il duplicato

    # Aggiungi il documento se non è duplicato
    db.add_documents([{"page_content": content, "metadata": metadata}])
    print(f"Documento aggiunto: {metadata['title']}")

# Funzione per caricare e aggiungere libri
def load_and_add_books(file_paths, db):
    """
    Carica i file di testo, genera i metadati e aggiunge i contenuti al database Chroma.
    """
    for file_path in file_paths:
        with open(file_path, "r", encoding="utf-8") as file:
            content = file.read()

        # Genera metadati per il libro
        metadata = {
            "title": file_path.split("/")[-1],  # Nome del file come titolo
            "author": "Unknown",  # Puoi specificare l'autore se noto
            "category": "Literature",  # Categoria generale
        }

        # Aggiungi al database evitando duplicati
        add_document_to_chroma(content, metadata, db)

# Esempio: elenco di file da caricare
file_paths = [
    "./divina_commedia.txt",
    "./promessi_sposi.txt",
    "./orlando_furioso.txt",
]

# Carica e aggiungi i libri al database
load_and_add_books(file_paths, db)

# Persisti il database
db.persist()

# Funzione per generare una risposta basata sul contesto
def get_chat_response(query, history):
    # Cerca nel Chroma DB
    results = db.similarity_search_with_score(query, k=5)
    context_text = "\n\n---\n\n".join([doc.page_content for doc, _score in results])

    # Costruisci la prompt
    prompt_template = ChatPromptTemplate.from_template(PROMPT_TEMPLATE)
    prompt = prompt_template.format(context=context_text, question=query)

    # Genera la risposta con il modello Hugging Face
    model = LLM()
    response_text = model.generate_response(prompt)

    # Aggiungi la risposta alla cronologia
    history.append(AIMessage(content=response_text))
    return response_text

# Interfaccia Gradio
def predict(message, history):
    history_langchain_format = []

    # Converti la cronologia per LangChain
    for msg in history:
        if msg['role'] == "user":
            history_langchain_format.append(HumanMessage(content=msg['content']))
        elif msg['role'] == "assistant":
            history_langchain_format.append(AIMessage(content=msg['content']))

    history_langchain_format.append(HumanMessage(content=message))

    # Ottieni la risposta dal sistema
    response = get_chat_response(message, history_langchain_format)

    return response

# Contenuto introduttivo
intro_content = """
# Book Query Bot
This bot allows you to search through a collection of books and find relevant information based on your query.
You can ask questions like:
- "What is the story of the Divine Comedy?"
- "Tell me about Don Abbondio in The Betrothed."
"""

# Costruisci l'interfaccia Gradio
with gr.Blocks() as demo:
    gr.Markdown(intro_content)
    chatbot = gr.ChatInterface(predict, type="messages")

# Avvia l'interfaccia
demo.launch(server_name="0.0.0.0", server_port=8080)
```
